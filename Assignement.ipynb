```python
q1=int(df.describe()["Exam_Score"]["25%"])
```

```python
q3=int(df.describe()["Exam_Score"]["75%"])
```

```python
IQR=q3-q1
print(IQR)
```

```python
lower_limit =q1-1.5*IQR
upper_limit =q3+1.5*IQR
```

```python
lower_limit
```

```python
df[(df["Exam_Score"]>upper_limit)    |  (df["Exam_Score"]<lower_limit)].count()
```

```python
# removing the outliers(trimming)
df[(df["Exam_Score"]<upper_limit) & (df["Exam_Score"]>lower_limit)].count()
```
